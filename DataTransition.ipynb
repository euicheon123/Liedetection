```python
import os
import pickle

# Path to the directory containing the combined poly data files
data_dir = r'C:\Users\User\Documents\Lie detect data\npy 데이터3'

# Ensure the output directory exists
output_dir = r'C:\Users\User\Documents\Lie detect data\TestData'
os.makedirs(output_dir, exist_ok=True)

# Initialize counter for file naming
counter = 1

# Iterate through all files in the data directory
for file_name in os.listdir(data_dir):
    if file_name.endswith('.pkl') and 'quer' in file_name:
        file_path = os.path.join(data_dir, file_name)

        # Load the combined Poly data file
        with open(file_path, 'rb') as file:
            combined_data = pickle.load(file)

        # Debug: Print the file being processed
        print(f"Processing file: {file_name}, Number of epochs: {len(combined_data)}")

        # Split and save individual epochs
        for i, epoch in enumerate(combined_data):
            individual_file_path = os.path.join(output_dir, f'eeg_quer_{counter}.pkl')
            with open(individual_file_path, 'wb') as individual_file:
                pickle.dump(epoch, individual_file)
                
            # Debug: Print each file being saved
            print(f"Saved epoch {i+1} as {individual_file_path}")
            
            counter += 1


print(f"Saved {counter-1} epochs.")

```
