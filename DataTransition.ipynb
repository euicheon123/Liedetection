```python
import os
import pickle

# Path to the directory containing the combined poly data files
data_dir = r'C:\Users\User\Documents\Lie detect data\npy 데이터2'

# Ensure the output directory exists
output_dir = r'C:\Users\User\Documents\Lie detect data\6M_PolyData'
os.makedirs(output_dir, exist_ok=True)

# Initialize counters for naming
truth_counter = 1
lie_counter = 1

# Iterate through all files in the data directory
for file_name in os.listdir(data_dir):
    if file_name.endswith('.pkl'):
        file_path = os.path.join(data_dir, file_name)

        # Load the combined Poly data file
        with open(file_path, 'rb') as file:
            combined_data = pickle.load(file)

        # Debug: Print the file being processed
        print(f"Processing file: {file_name}, Number of epochs: {len(combined_data)}")

        # Determine if the file contains truth or lie data based on the filename
        if 'truth' in file_name and 'ansr' in file_name and 'poly' in file_name:
            label = 'truth'
            counter = truth_counter
        elif 'lie' in file_name and 'ansr' in file_name and 'poly' in file_name:
            label = 'lie'
            counter = lie_counter
        else:
            continue

        # Split and save individual epochs
        for i, epoch in enumerate(combined_data):
            individual_file_path = os.path.join(output_dir, f'poly_ansr_{label}_{counter}.pkl')
            with open(individual_file_path, 'wb') as individual_file:
                pickle.dump(epoch, individual_file)
                
            # Debug: Print each file being saved
            print(f"Saved {label} epoch {i+1} as {individual_file_path}")
            
            counter += 1

        # Update the respective counter
        if label == 'truth':
            truth_counter = counter
        elif label == 'lie':
            lie_counter = counter

print(f"Saved {truth_counter - 1} truth epochs and {lie_counter - 1} lie epochs.")

```
